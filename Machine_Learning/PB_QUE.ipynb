#### Predict the y values for x = np.arange(5). (Don't split data for training/testing)

```python
df = pd.read_csv("DATAS/datasets/iris.csv")
df.head()
```

```python
x = df.drop(["Species","Id"] , axis=1)
y = df["Species"]

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
KNN = KNeighborsClassifier(n_neighbors=3)
KNN.fit(X_train , y_train)

pre = KNN.predict(X_test)
print(accuracy_score(y_test , pre))
print(confusion_matrix(y_test , pre))
```

```python
df = pd.read_csv("Datasets/petrol_consumption.csv")
print(df.head())

x = df.drop('Petrol_Consumption' , axis=1)
y = df['Petrol_Consumption']

X_train , X_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=42)

model = LinearRegression()
model.fit(X_train , y_train)
y_pre = model.predict(X_test)

r2 = r2_score(y_test , y_pre)
print("R2 score :" , r2)

mse = mean_squared_error(y_test , y_pre)
print("MSE :" , mse )

print("---- co-efficients ---- ")
for name , coef in zip(X_train.columns , model.coef_) :
    print(f" {name} --> {coef} ")

print("---> intercept :" , model.intercept_)
```

```python
df = pd.read_csv("Datasets/RealEstate.csv")

df = df.drop("No" , axis=1)
print(df.head(2))

print("Total null values :",df.isna().sum().sum())

x = df.drop("Y house price of unit area" , axis=1)
y = df['Y house price of unit area']

X_train , X_test , y_train , y_test = train_test_split(x , y , test_size=0.3 , random_state=110)

model = LinearRegression()
model.fit(X_train , y_train)
y_pre = model.predict(X_test)

r2 = r2_score(y_test , y_pre)
print("\n--> R2 score :" , r2)

print("\n --- coef --- \n")
for name , coef in zip(X_train.columns , model.coef_) :
    print(f" {name} --> {coef} ")

print("\n--> Intercept :" , model.intercept_)
```

```python
df = pd.read_csv("Datasets/winequality.csv")
print(df.head(2))
df = df.drop('Id' , axis=1)
print("total null values :",df.isna().sum().sum())

x = df.drop('quality' , axis=1)
y = df['quality']

X_train , X_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state=42)

model = LinearRegression()
model.fit(X_train , y_train)
y_pre = model.predict(X_test)

r2 = r2_score(y_test , y_pre)
print("r2 score :" , r2)

mse = mean_squared_error(y_test , y_pre)
print("MSE : " , mse)

print("\n --- coef ---\n")
for name , coef in zip(X_train.columns , model.coef_) :
    print(f" {name} --> {coef} ")

print("\n intercept :" , model.intercept_)

# predict data
user_data = [[8,0.4,0.4,15,0.048,40,150,0.99,3,0.45,10.5]]
user_pre = model.predict(user_data)
print("\nQuality for user data :" , user_pre[0])
```
