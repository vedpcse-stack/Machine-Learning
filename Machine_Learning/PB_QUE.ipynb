```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import r2_score , accuracy_score , confusion_matrix
from sklearn.preprocessing import PolynomialFeatures 
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
```

## PB : 219
#### For x = np.arange(0, 30) and y = np.array([3, 4, 5, 7, 10, 8, 9, 10, 10, 23, 27, 44, 50, 63, 67, 60, 62, 70, 75, 88, 81, 87, 95, 100, 108,
#### 135, 151, 160, 169, 179]), apply polynomial regression using scikit learn library and calculate R squared, coeficient and intercept.
#### Predict the y values for x = np.arange(5). (Don't split data for training/testing)

```python
x = np.arange(0, 30)
y = np.array([3, 4, 5, 7, 10, 8, 9, 10, 10, 23, 27, 44, 50, 63, 67, 60, 62, 70, 75, 88, 81, 87, 95, 100, 108,
135, 151, 160, 169, 179])

df_x = pd.DataFrame(x)

poly = PolynomialFeatures(degree=3)
df_x = poly.fit_transform(df_x)
test_x = poly.fit_transform( np.arange(5).reshape(-1,1) )
model = LinearRegression()
model.fit(df_x,y)

# pred = model.predict(df_x)
# print(pred)

pre_ans = model.predict(test_x)
print( "Predict : ", pre_ans)
print("coff" , model.coef_)
print("intercept" , model.intercept_)
```

## PB : 217
#### Given a real estate price size year dataset, implement multiple linear regression using scikitlearn library. Using the model, make
#### a prediction about an apartment price with size 750 sq.ft. for 2009.Also Calculate R squared, coeficient and intercept. (Don't split
#### data for training/testing)

```python
df = pd.read_csv("DATAS/Datasets/real_estate.csv")

x = df.drop("price" , axis=1)
y = df["price"]

model = LinearRegression()
model.fit(x , y)

pre = model.predict([[750,2009]])
print("Pre :",pre)
print("R squared :" , model.score(x,y))
```

Split the dataset into features (X) and labels (y).
Split the features and labels into training and testing sets with a test size of 0.2 and a random state of 42.
Initialize a kNN classifier with 3 neighbors.
Train the classifier on the training set.
Make predictions on the test set.
Calculate and print the accuracy score of the classifier

```python
df = pd.read_csv("DATAS/datasets/iris.csv")
df.head()
```

```python
x = df.drop(["Species","Id"] , axis=1)
y = df["Species"]

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
KNN = KNeighborsClassifier(n_neighbors=3)
KNN.fit(X_train , y_train)

pre = KNN.predict(X_test)
print(accuracy_score(y_test , pre))
print(confusion_matrix(y_test , pre))
```
