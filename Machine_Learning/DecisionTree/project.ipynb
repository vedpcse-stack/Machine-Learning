```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report , accuracy_score , confusion_matrix , recall_score , precision_score
```

```python
df = pd.read_csv("../Loan.csv")
print(df.head())
print(df.shape)
print(df.columns)
print(df.info())
```

```python
df.isna().sum()
```

```python
df = df.fillna(value={
    "LoanAmount" : df["LoanAmount"].median(),
    "Income" : df["Income"].median()
})
df.isna().sum()
```

```python
df.duplicated()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

```python
x = df.drop("LoanApproved" , axis=1)
y = df["LoanApproved"]
```

```python
x = pd.get_dummies(x , columns=["Education" , "SelfEmployed" , "PropertyArea" , "PreviousLoans" ] , drop_first=True)
x[["LoanAmount","Income"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
KNN = KNeighborsClassifier(n_neighbors=3)
KNN.fit(X_train , y_train)
Pre = KNN.predict(X_test)

acc = accuracy_score(y_test , Pre)
recall = recall_score(y_test , Pre , pos_label="Yes")
precision = precision_score(y_test , Pre , pos_label="Yes")
print(acc)
print(recall)
print(precision)
```

```python
Tree = DecisionTreeClassifier()
Tree.fit(X_train , y_train)
Pre1 = Tree.predict(X_test)

acc1 = accuracy_score(y_test , Pre1)
recall1 = recall_score(y_test , Pre1 , pos_label="Yes")
precision1 = precision_score(y_test , Pre1 , pos_label="Yes")
print(acc1)
print(recall1)
print(precision1)
```
