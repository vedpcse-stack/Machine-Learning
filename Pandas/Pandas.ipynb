# Group By

```python
data = {
    'category' : [ 'a' , 'b' , 'a' , 'b' , 'a' , 'b' , 'a' , 'b' ],
    'store' : [ 's1' , 's1' , 's2' , 's2' , 's1' , 's2' , 's2' , 's1' ],
    "sales" : [ 100 , 200 , 150 , 250 , 120 , 180 , 200 , 300 ],
    "quantity" : [10,15,12,18,8,20,15,25],
    "date" : pd.date_range( '2025-01-01' , periods=8)
}

df = pd.DataFrame(data)
df
```

```python
# group by category and cal. tha sum of sales

df.groupby( "category" )
```

```python
cat = df.groupby( "category" )

for i , v in cat :
    print(i)
    print(v)
```

```python
cat = df.groupby( "category" )[ "sales"].sum()
cat
```

```python
# group by Store & cal. sum of sales

store = df.groupby( "store" )["sales"].sum()

store
```

```python
# groueby multiple columns

store = df.groupby( [ 'category' , 'store' ] )["sales"].sum()
print(store)
```

```python
store = df.groupby( [ 'store' , 'category'] )["sales"].sum()
print(store)
```

# Pivot Tables

```python
data = {
    "date" : pd.date_range( '2025-01-01' , periods=20 ),
    "product" : ['a' , 'b' , 'c' , 'd'] * 5,
    "region" : [ 'east' , 'west' , 'north' , 'south' ] * 5 ,
    "sales" : np.random.randint(100 , 1000 , 20),
    "units" : np.random.randint(10,100,20),
    "rep" : [ 'john' , 'mary' , 'bob' , 'alice' ] * 5
}

df = pd.DataFrame(data)
df["month"] = df['date'].dt.month_name()
df["quarter"] = 'Q' + df["date"].dt.quarter.astype(str)

df
```

```python
pd.pivot_table(df , values="sales" , index="region" , columns= "product" ) # if multiple vales so give mean value for that cell
```

```python
pd.pivot_table(df , values="sales" , index="region" , columns= "product" , aggfunc="min" ) 
```

```python
pd.pivot_table(df , values="sales" , index="region" , columns= "product" , aggfunc="median" ) 
```

```python
pd.pivot_table( df , values=[ "sales" , "units" ] , columns="product" , index="region" )
```

```python
df = pd.read_csv(r"anime.csv")
df
```

```python
# make new column for ep. count

def countEp(txt) :
    temp = txt.split("(")

    ep = temp[1].split()

    ep = ep[0]

    return int(ep)

df["ep_count"] = df["Title"].apply(countEp)

df
```

```python
# make column for timestamp

def timeStamp(txt) :
    temp = txt.split(")")
    time = temp[1][:19]
    return time

df["timestamp"] = df["Title"].apply(timeStamp)
df
```

```python
# find total months
df['total_months'] = (
    pd.to_datetime(df['timestamp'].str.split(' - ').str[1], format='%b %Y').dt.to_period('M') -
    pd.to_datetime(df['timestamp'].str.split(' - ').str[0], format='%b %Y').dt.to_period('M')
).apply(lambda x: x.n +1)

df
```

```python
# give me top highest Scoring anime

df["Title"].head()
```

```python
# highest ep. count

df[ df["ep_count"] == df["ep_count"].max() ]["Title"]
```
