```python
import networkx as nx
```

```python
instagram = {'person1': [0,1,1,0,1], 'person2': [0,0,1,0,1], 'person3': [1,1,0,1,1],'person4': [1,1,1,0,0], 'person5': [1,1,0,0,0]}
```

```python
g = nx.DiGraph()
```

```python
g.add_nodes_from(instagram.keys())

for i,j in instagram.items() :
    for k,l in enumerate(j) :
        if l == 1 :
            g.add_edge(i,f"person{k+1}")
            

nx.draw_networkx(g)
plt.show()
```
