```python
import numpy as np
```

```python
arr = np.array( [2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201] )
```

```python
Q1 = np.percentile(arr , 25)
Q3 = np.percentile(arr , 75)

Q2 = np.percentile(arr , 50)

print(Q1 , Q2 , Q3)
```

```python
IQR = Q3 - Q1

Upper_fance = Q3 + ( 1.5 * IQR )
Lower_fance = Q1 - ( 1.5 * IQR )


print(Upper_fance , Lower_fance)
```

```python
new_list = []

for i in arr :
    if i >= Lower_fance and i <= Upper_fance :
        new_list.append( int(i) )
    
print(new_list)

arr2 = np.array( new_list )

print(arr2)
```
