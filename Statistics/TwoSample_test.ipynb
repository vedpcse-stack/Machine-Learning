```python
import numpy as np
from scipy import stats
```

# T test

```python
group_a = np.array( [85 , 88 , 90 , 92 , 87 , 85 , 89 , 91 , 86 , 88] )
group_b = np.array( [82 , 84 , 80 , 83 , 81 , 79 , 78 , 85 , 84 , 83] )
```

```python
a_mean = group_a.mean()
b_mean = group_b.mean()

n1 = len(group_a)
n2 = len(group_b)

a_std = group_a.std()
b_std = group_b.std()
```

```python
t_value , p_value = stats.ttest_ind(group_a , group_b ,  equal_var=False)
```

```python
print(t_value)
print(p_value)
```

```python
alpha = 0.05

if p_value < alpha :
    print( "i will reject the null hypothesis" )
else:
    print( "i will accept the null hypothesis" )
```
