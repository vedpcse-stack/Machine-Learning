```python
import numpy as np
from scipy import stats
```

```python
sample = np.array([172,174,168,169,171,173,175,170,169,172])
```

```python
sample_mean = sample.mean()
sample_std = sample.std(ddof=1)  # ddof = degree of freedom
n = len(sample)

population_mean = 170
```

```python
t_score = ( sample_mean - population_mean ) / ( sample_std / np.sqrt(n) )

print(t_score)
```

```python
p_value = 2 * ( 1 - stats.t.cdf(abs(t_score) , df = n - 1) )

print(p_value)
```

```python
alpha = 0.05

if p_value < alpha :
    print( "i will reject the null hypothesis" )
else:
    print( "i will accept the null hypothesis" )
```
