```python
import numpy as np
from scipy.stats import norm
```

```python
sample = np.array([172,174,168,169,171,173,175,170,169,172])
```

```python
population_mean = 170
population_std = 3
```

```python
sample_mean = sample.mean()
n = len(sample)
```

```python
z_score = ( sample_mean - population_mean ) / ( population_std / np.sqrt(n) )

print(z_score)
```

```python
p_value = 2 * ( 1 - norm.cdf( z_score ) )

print(p_value)
```

```python
alpha = 0.05

if p_value < alpha :
    print( "i will reject the null hypothesis" )
else:
    print( "i will accept the null hypothesis" )
```
