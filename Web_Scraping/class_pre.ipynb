```python
import requests
from bs4 import BeautifulSoup

def get_weather(city):
    # Google weather search URL
    url = f"https://www.google.com/search?q=weather+{city}"

    headers = {
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/124.0.0.0 Safari/537.36"
        )
    }

    response = requests.get(url, headers=headers)

    if response.status_code != 200:
        print("Failed to fetch data")
        return

    soup = BeautifulSoup(response.text, "html.parser")

    try:
        location = soup.find("div", id="wob_loc").text
        time = soup.find("div", id="wob_dts").text
        condition = soup.find("span", id="wob_dc").text
        temperature = soup.find("span", id="wob_tm").text

        print("Location :", location)
        print("Time     :", time)
        print("Condition:", condition)
        print("Temp     :", temperature + "°C")

    except AttributeError:
        print("Could not fetch weather data")

# Example
city_name = input("Enter city name: ")
get_weather(city_name)
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests

url = "https://www.politifact.com/factchecks/list/"

data = requests.get(url)

html = data.text

soup = BeautifulSoup(html , "html.parser")

box = soup.find_all("li" , class_="o-listicle__item")

name = []
date = []
desc = []
for i in box :
    name.append( i.find( "a" , class_="m-statement__name").text.strip() )
    date.append( i.find( "footer" , class_="m-statement__footer").text.strip() )
    desc.append( i.find( "div" , class_="m-statement__desc").text.strip() )

auth = []
date_fatch = []

for i in date :
    temp = i.split("•")
    auth.append(temp[0].strip())
    date_fatch.append(temp[1].strip())

df = pd.DataFrame({
    "name" : name ,
    "auth" : auth,
    "date" : date_fatch,
    "Desc" : desc
})

df.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

with open("flipkart.html" , encoding="utf-8") as f :
    html = f.read()

soup = BeautifulSoup(html)

mobile = soup.find_all("div" , class_ = "_3pLy-c")

rating = []
ram = []
battery = []
name = []

with_dis = []
org = []

total_dis = []

t1 = []
t2 = []

for i in mobile :
    rating.append( i.find( "div" , class_="_3LWZlK").text.strip() )
    ram.append( i.find( "li" , class_="rgWa7D").text.strip() )
    name.append( i.find( "div" , class_="_4rR01T").text.strip() )

    temp = i.find_all("li" , class_="rgWa7D")[3].text
    battery.append(temp)

    dis_pre = i.find("div" , class_="_30jeq3 _1_WHN1")
    if dis_pre :
        with_dis.append(dis_pre.text)
        temp1 = dis_pre.text[1:].replace("," , "" )
    else :
        temp1 = 0
        with_dis.append(0)
    t1.append(temp1)

    org_pre = i.find("div" , class_="_3I9_wc _27UcVY")
    if org_pre :
        org.append(org_pre.text)
        temp2 = org_pre.text[1:].replace("," , "")

    else :
        temp2 = 0
        org.append(0)
    t2.append(temp2)
    
    total_dis.append( float(int(temp2) - int(temp1)) )

df = pd.DataFrame({
    "name" : name,
    "ram" : ram,
    "battery" : battery,
    "rating" : rating,
    "discount price" : with_dis,
    "org price" : org,
    "dis" : total_dis
})

print("t1" , t1)
print("t2" , t2)

df
```
