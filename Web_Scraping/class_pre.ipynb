```python
from bs4 import BeautifulSoup

with open("WS.html") as f :
    html = f.read()

soup = BeautifulSoup(html ,"html.parser")

# z = soup.find("h1")
# print(z)
# print(z.text)

# print('===============')

# p = soup.find("p")
# print(p)
# print(p.text)

# print('===============')

# all = soup.findAll("ul")
# print(all)
# print(all[0].text)

print('===============')

# s1 = soup.findAll("div" , {"class" : "student"})
# 

# df = pd.DataFrame([
#     s1[0].text.strip().split("\n")
# ],columns=["ID" , "name" , "Age" , "Course" ])  
# df

import pandas as pd
print('===============')
l1 = soup.find_all("div" , class_ = "student" )
name = []
age = []
course = []
link = []
email_name = []
ext = []
img = []
for i in l1 :
    name.append(i.find("p" , class_ = "name").text.strip())
    age.append(i.find("p" , class_ = "age").text.strip())
    course.append(i.find("p" , class_ = "course").text.strip())
    l = i.find("a")
    link.append(l["href"])
    
    e = i.find("p" , class_ = "email").text.split("@")
    email_name.append(e[0])
    ext.append("@"+e[1])

    img.append( i.find("img")["alt"]  )

df = pd.DataFrame({
    "name" : name ,
    "age" : age,
    "course" : course,
    "link" : link,
    "email name" : email_name,
    "ext" : ext,
    "image" : img
})
df
```
