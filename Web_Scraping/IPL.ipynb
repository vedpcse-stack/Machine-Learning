```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url="https://www.iplt20.com/auction"
data = requests.get(url)

soup = BeautifulSoup(data.text , "html.parser")

x = soup.find("table",class_="ih-td-tab w-100 auction-tbl")
rows = x.find_all("tr")

cols = [i.text.strip() for i in rows[0].find_all("th")][2:]

data = []

for i in rows[1:] :
    data.append([j.text.strip() for j in i.find_all("td")][1:])

df = pd.DataFrame(data , columns=cols , index=range(1,11))

df
```
