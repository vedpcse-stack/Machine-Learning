```python
from bs4 import BeautifulSoup
import pandas as pd

with open(r"html files/Scrape_this.html") as f :
    html = f.read()

soup = BeautifulSoup(html, "html.parser")

table = soup.find("table" , class_ = "common-table")

rows = table.find_all("tr")[1:]

data = []

for i in rows :
    temp = [ j.text.strip() for j in i.find_all("td")]
    link = i.find("td").find("a")["href"]

    data.append([temp[0] , link , temp[1] , temp[5] , temp[6]] )

df = pd.DataFrame(data , columns=["company name" , "company Link" , "current price" , "Value" , "Volume"])

print(df.head())
print(df.shape)

df.dropna()
print(df.shape)

```
