```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = "https://www.cricbuzz.com/cricket-series/9241/indian-premier-league-2026/points-table"

data = requests.get(url)
html = data.text

soup = BeautifulSoup(html , "html.parser")

all_team = soup.find_all("div" , class_ = "point-table-grid")[1:]

team_curr_rank = []
team_name = []
team_total_match = []
team_win = []
team_loss = []
team_NR = []
team_PTS = []
team_NRR = []

for i in all_team :
    team_curr_rank.append(int(i.find_all("div" , class_="text-cbPointsTableTextGrey")[0].text))
    team_name.append(i.find_all("div" , class_="text-cbPointsTableTextGrey")[1].text)
    team_total_match.append(int(i.find_all("div" , class_="text-cbPointsTableTextGrey")[2].text))
    team_win.append(int(i.find_all("div" , class_="text-cbPointsTableTextGrey")[3].text))
    team_loss.append(int(i.find_all("div" , class_="text-cbPointsTableTextGrey")[4].text))
    team_NR.append(int(i.find_all("div" , class_="text-cbPointsTableTextGrey")[5].text))
    team_PTS.append(int(i.find("div" , class_="text-gray-800").text))
    team_NRR.append(float(i.find_all("div" , class_="text-cbPointsTableTextGrey")[6].text))


df = pd.DataFrame({
    "TEAMS" : team_name,
    "P" : team_total_match,
    "W" : team_win,
    "L" : team_loss,
    "NR" : team_NR,
    "PTS" : team_PTS,
    "NRR" : team_NRR
},index=team_curr_rank)
df
```
