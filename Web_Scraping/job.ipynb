```python
from bs4 import BeautifulSoup
import pandas as pd , requests
```

```python
with open("job_page.html") as f :
    html = f.read()

soup = BeautifulSoup(html ,"html.parser")

div = soup.find_all("div" , class_ = "job_seen_beacon")

name = []
title = []
place = []
sal = []
for i in div :
    name.append(i.find("a" , class_ = "jcs-JobTitle").text)
    title.append(i.find("div" , class_="css-1afmp4o e37uo190").text)
    place.append(i.find("div" , class_="eu4oa1w0").text)
    x = i.find("div" , class_="css-18z4q2i eu4oa1w0")
    if x :
        sal.append(x.text.replace("â‚¹" , ""))
    else :
        sal.append(None)

df = pd.DataFrame({
    "name" : name,
    "title" : title,
    "place" : place,
    "salary" : sal
})

df
```
