```python
from bs4 import BeautifulSoup
import pandas as pd

with open(r"html files/Popular_Quotes.html") as f :
    html = f.read()

soup = BeautifulSoup(html , "html.parser")

#! 1. h1 tag
print(soup.find("h1").text)

#! 2. title
print("title :",soup.find("title").text[::-1])

#! 3. no. of script tags
print("no. of script tags :",len(soup.find_all("script")))
print("-------------")


data = soup.find_all("div" , class_="quoteText")
Quotes = []
Author = []

#! 4. listed tags under first Author
x = soup.find("div" ,class_ = "quoteDetails").find("div" , class_ = "quoteFooter").find_all("a")
print("Tags under Author Oscar : \n")
for i in x :
    print(i.text)

print("\n---------------------")
#! 5. Qoutes & author
for i in data[:5] :
    temp = i.text.split("―")
    print("Quote :" , temp[0].strip())
    print("Author :" , temp[1].strip())
    



#! 6. Quote contain "accomplishment"
for i in data :
    temp = i.text.split("―")
    Quotes.append(temp[0].strip())
    Author.append(temp[1].strip())
    if "accomplishment" in temp[0].strip() :
        print("Quote :" , temp[0].strip())

#! 7. likes
likes = []
like = soup.find_all("a" , class_ = "smallText")
for i in like :
    likes.append(int(i.text.strip().split(" ")[0]))
likes.sort()
print("----------")
print("likes :" , likes)
print("-----------")

#! 8. multi Quotes
print("----------")
multi_qoutes_author = []
for i in Author :
    if Author.count(i) > 1 :
        multi_qoutes_author.append(i)
print("Authors with multi Quotes :" , set(multi_qoutes_author))

#! 9. create DataFrame
df = pd.DataFrame({
    "Quotes" : Quotes,
    "Author" : Author,
    "Likes" : likes
})

df
```
